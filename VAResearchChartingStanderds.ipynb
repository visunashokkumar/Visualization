```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
start = '1998-01-01'
end = '2023-04-04'
df = yf.download('^BSESN', start=start, end=end)
```

```python
df = df['Adj Close']
```

```python
# Create a figure and axis object
fig, ax = plt.subplots(figsize=(16,3))

# Defining the color scheme
bgColor = '#000000'
spColor = '#ffa500'

# Plot the data as a line chart
ax.plot(df, linewidth=2, color='#D6230D')

# Set the x and y axis labels and limits
ax.set_xlabel('Date', fontsize=10, color=spColor)
ax.set_ylabel('Price', fontsize=10, color=spColor)
ax.set_xlim(df.index[0],)

# Set the chart title
ax.set_title('Stock Price', fontsize=12, fontweight='bold', color=spColor)

# Add a grid
ax.grid(axis='y', color=spColor, linestyle='-', linewidth=0.25)

# Remove the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set the tick parameters
ax.tick_params(axis='both', which='both', labelsize=9, length=4, color=spColor, labelcolor=spColor)

# Set the background color to black
fig.patch.set_facecolor(bgColor)
ax.set_facecolor(bgColor)

# Set the color of the x and y axis lines to Bloomberg orange
ax.spines['bottom'].set_color(spColor)
ax.spines['left'].set_color(spColor)

# Display the chart
plt.show()
```
